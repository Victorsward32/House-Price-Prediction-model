```python
data.drop(columns=['Gymnasium', 'Lift Available', 'Car Parking', 'Maintenance Staff', '24x7 Security', "Children's Play Area", "Clubhouse", "Intercom", "Landscaped Gardens", "Indoor Games", "Gas Connection", "Jogging Track", "Swimming Pool"], inplace=True)
```

```python
data['price_per_sqft'] = data['Price'] / data['Area']
```

```python
data['price_per_sqft']
```

```python
data['Location']
```

```python
data['Location'] = data['Location'].apply(lambda x: x.strip())
location_count = data['Location'].value_counts()
```

```python
location_count_less_10 = location_count[location_count <= 10]
location_count_less_10
```

```python
data['Location'] = data['Location'].apply(lambda x: 'other' if x in location_count_less_10 else x)
```

```python
data['Location'].value_counts()
```

```python
data.describe()
```

```python
data.info()
```

```python
def remove_outliers_sqft(df):
  df_output = pd.DataFrame()
  for key, subdf in df.groupby('Location'):
    m = np.mean(subdf.price_per_sqft)
    st = np.std(subdf.price_per_sqft)
    
    gen_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m+st))]
    df_output = pd.concat([df_output, gen_df], ignore_index = True)
  return df_output
data = remove_outliers_sqft(data)
data.describe()
```

```python
data.shape
```

```python
data.drop(columns=['price_per_sqft','New/Resale', 'Unnamed: 0'], inplace=True)
```

```python
data.head()
```

```python
data.to_csv('/content/drive/MyDrive/Cleaned_data.csv')
```

```python
X = data.drop(columns=['Price'])
y = data['Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

#**Applying Linear Regression**

```python
column_trans = make_column_transformer((OneHotEncoder(sparse=False), ['Location']),remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression(normalize = True)
```

```python
pipe = make_pipeline(column_trans, scaler, lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_lr)
```

#**Applying Lasso**

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans, scaler, lasso)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test, y_pred_lasso)
```

#**Applying Ridge**

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans, scaler, ridge)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test, y_pred_ridge)
```

```python
print("No Regularization: ", r2_score(y_test, y_pred_lr))
print("Lasso: ", r2_score(y_test, y_pred_lasso))
print("Ridge: ", r2_score(y_test, y_pred_ridge))
```
